### Define a function for displaying images

* We will want to look at images as we examine and analyze the data, so here will will define a function for displaying images. 
* This will handle getting the stretch of the image correct. 
* Feel free to play with the settings and colormap once you get the hang of the process.



```python
def imdisplay(image, v1perc=10, v2perc=95):
    '''
    display an image 
    OPTIONAL KEYWORD PARAMETERS
    v1perc: one end of the colormap assigned to the v1perc percent lowest flux 
    v2perc: the other end of the colormap assigned to the v2perc percent highest flux    
    '''
    # make sure image is an np array
    nimage = np.array(image)
    # determine the pixel values at the 10th and 95th percentile
    v1 = scoreatpercentile(nimage,v1perc)
    v2 = scoreatpercentile(nimage,v2perc)
    # display using imshow
    #
    # you can play with alternate cmaps in the function below, such as "viridis" or "gray"
    # The 'gray_r' color map reverses the color-scale so that dark display pixels are the brightest in the image
    #
    # vmin and vmax set the min and max pixel values that
    # will be mapped to the extremes of the colormap
    plt.imshow(nimage,cmap='gray_r',vmin=v1,vmax=v2)
    plt.colorbar(fraction=.08)
```

We want to find the **gain** $=g$ where the gain is defined by $N_{ADU} = \frac{N_e}{g}$ where $N_e = $ the number of actual electrons generated by light.

We start by defining $N_c \equiv $ The total number of counts in our flat field [ADU]

$N_{light} \equiv $ The number of counts resulting from the collection of light [ADU]

$N_{bias} \equiv $ The number of counts from the bias [ADU]

$N_c = N_{light} + N_{bias}$ shows what happens when we take data.  

Using propogation of error: $\sigma_{c}^2 = \left(\frac{\partial N_{c}}{\partial N_{light}}\right)^2\sigma_C^2 + \left(\frac{\partial N_{c}}{\partial N_{bias}}\right)^2\sigma_{bias}^2$ . 


To get the number of counts just from the collection of light you have: $N_{light} = N_c - N_{bias}$


$\sigma_{c}^2= \sigma_{light}^2 + \sigma_{bias}^2$.  Therefore  $\sigma_{light}^2= \sigma_c^2 - \sigma_{bias}^2$.

Remembering from above that $N_{light} = \frac{N_e}{g}$ we can use propogation of error to define 

$\sigma_{light}^2 = \left(\frac{\partial N_{light}}{\partial N_e}\right)^2\sigma_e^2 + \left(\frac{\partial N_{light}}{\partial g}\right)^2\sigma_g^2$.  

The second term disappears because we assume $\sigma_g=0$.

We then have $\sigma_{light}^2 = \frac{\sigma_e^2}{g^2}$ and from Poisson statistics $\sigma_e^2 = N_e$.

Rearranging the firsdt term we have 

$g^2 = \frac{\sigma_e^2}{\sigma_{light}^2} = \frac{N_e}{\sigma_{light}^2} = \frac{g N_{light}}{\sigma_{light}^2}$

Where $\sigma_{light}$ is what we actually measure as the standard deviation from our image.

We therefore get $g = \frac{N_{light}}{\sigma_c^2 - \sigma_{bias}^2} = \frac{N_c - N_{bias}}{\sigma_c^2 - \sigma_{bias}^2}$.

Compared to the actual gain we are within 20%, which is pretty good.
